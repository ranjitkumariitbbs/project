**1. Logistic Regression**

Predicted purchased decision of a customer based on salary and age parameter

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import dataset
dataset=pd.read_csv('/content/drive/MyDrive/Dataset/Social_Network_Ads.csv')
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
# Splitting the dataset into training set and test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
# Features scaling (optional): in order to improve the performance
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

Training the Logistic regression model on Training set 

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,Y_train)


```

Predict the Test set result


```python
Y_pred=classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1),Y_test.reshape(len(Y_test),1)),1))

```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
# Accuracy 
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```

**Precision, Recall and F1 score**

```python
from sklearn.metrics import precision_score , recall_score, f1_score
# Precision Score
precision_score(Y_test,Y_pred)
```

```python
# Recall Score
recall_score(Y_test, Y_pred)
```

```python
# F1 Score
f1_score(Y_test, Y_pred)
```

**Precision, Recall curve**

```python
from sklearn.metrics import precision_recall_curve
precision,recall,thersholds = precision_recall_curve(Y_test, Y_pred)
```

**2. K Nearest Neighbors (KNN)**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)

classifier.fit(X_train,Y_train)

# Predict the test set result

Y_pred=classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1),Y_test.reshape(len(Y_test),1)),1))

```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
# Accuracy 
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```

**3. Support Vector Machine (SVM)**

```python

# Training the SVM model on the Training set

from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(X_train,Y_train)

# Predicting a new result
print(classifier.predict(sc.transform([[1,820]])))

```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
# Accuracy 
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```
