```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
```

```python
dataset=pd.read_csv("/content/drive/MyDrive/Dataset/Bank_Customers_Dataset.csv")
```

```python
print(dataset.head())
```

```python
print(dataset)
```

```python
X=dataset.iloc[:,3:-1].values
Y=dataset.iloc[:,-1].values

print(X)
```

```python
print(Y)
```

**Encode Categorical Data**

```python
# Encode categorical data (Gender) using Label Encoder
# Since we have only two categorical data

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])
# Apply label encoding on all the row and only on column of index 2

print(X)
```

```python
# Encode Categorical data using OneHotEncoder
# Since we have more than two categorical data

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])] , remainder='passthrough')

X=np.array(ct.fit_transform(X))
print(X)

# When we perform OneHotEncoder then dumy variable move to first column of the matrix features
```

Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(X,Y,test_size=0.20, random_state=0)

```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

```

**Building the ANN**

Initializing the ANN

```python
ann=tf.keras.models.Sequential()
```

Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
# Hyper parameter value i.e number of neuron in hidden layer is random not a fixed or thumb rule
```

Adding the input layer and the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

**Training the ANN**

Compiling the ANN

```python
ann.compile(optimizer='adam', loss='binary_crossentropy' , metrics=['accuracy'] )
```

Training the ANN on Training set

```python
# Fitting the ANN to the Training set
ann.fit(x_train, y_train, batch_size= 32,epochs=100 )

```

**Making the predictions and evaluating the model**

Predicting the Test set results 

```python
y_pred= ann.predict(x_test)
y_pred=(y_pred>0.5)
```

Details of customer for prediction

```python
Country=1  # 0: ,1: ,2:
Credit_Score=950
Gender=1   # 0:Female, 1:Male
Age=40
Tenure=3
Balance=70000
Number_of_product=1
Has_CC=1   # 0: Has no Credit card, 1: Has Credit card
Is_Active_Customer=1   # 0:Not a active Customer, 1: Active Customer
Estimated_Salary=9000
```

New prediction

```python
new_prediction=ann.predict(sc.transform(np.array([[1,0,1, Credit_Score, Gender, Age,Tenure, Balance, Number_of_product, Has_CC, Is_Active_Customer, Estimated_Salary]])))
# Country: France: 1,0,1  Spain:0,0,1 Germany: 0,1,0

# new_prediction=(new_prediction>0.5)

print(new_prediction)

if new_prediction>0.5:
  print("Customer will stay in Bank")
else:
  print("Customer may leave the Bank")
```

**Confusion matrix**

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

**Accuracy**

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
